### Три графика — три вопроса
- **Гистограмма** → «сколько чего?» (распределение одного признака)
- **Scatter** → «связаны ли два признака?» (каждая точка — один объект)
- **Heatmap** → «как связаны все признаки сразу?» (цветом)

## Шаг 1. Подключаем инструменты для графиков
`matplotlib` рисует, `seaborn` делает это красиво.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")  # приятная тема

# знакомый датасет с прошлого урока
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
print("Данные загружены, рисуем!")
```

## График 1. Гистограмма — «сколько чего?»
Показывает, сколько пассажиров какого возраста было на борту.

```python
plt.figure(figsize=(8,4))
sns.histplot(data=df, x="Age", bins=20)
plt.title("Распределение возраста пассажиров")
plt.show()
```

**Вопрос:** в каком возрасте было больше всего пассажиров? Видно по самому высокому столбику.

## График 2. Scatter — «связаны ли два признака?»
Каждая точка — один пассажир. По осям: возраст и цена билета.

```python
plt.figure(figsize=(8,4))
sns.scatterplot(data=df, x="Age", y="Fare")
plt.title("Возраст и цена билета")
plt.show()
```

**Вопрос:** видишь ли ты чёткую форму (линию вверх/вниз) или точки разбросаны хаотично?

## График 3. Heatmap — связь ВСЕХ признаков сразу
Берём только числовые столбцы и считаем корреляцию.
🔴 красное — сильная связь, 🔵 синее — слабая или обратная.

```python
# оставляем только числовые столбцы
numeric = df.select_dtypes(include="number")

plt.figure(figsize=(8,6))
sns.heatmap(numeric.corr(), annot=True, cmap="coolwarm")
plt.title("Корреляция признаков")
plt.show()
```

**Вопрос:** найди самый «красный» квадрат (кроме диагонали — она всегда 1).
Какие два признака связаны сильнее всего?

## ⚠️ Важнейшее правило темы
**Корреляция — это НЕ причина.**

Летом растут продажи мороженого и растёт число утонувших. Связь есть? Да.
Мороженое топит людей? Нет — просто оба зависят от жары.

График показывает *связь*. Думать о *причине* должен человек.

## ⭐ Для тех, кто хочет больше
Раскрасим точки по выживанию — `hue` добавляет третий признак цветом.

```python
plt.figure(figsize=(8,4))
sns.scatterplot(data=df, x="Age", y="Fare", hue="Survived")
plt.title("Возраст, цена билета и выживание")
plt.show()
```

```python
# Сетка всех scatter сразу — одним взглядом видно все связи
sns.pairplot(df[["Age","Fare","Pclass","Survived"]])
plt.show()
```
