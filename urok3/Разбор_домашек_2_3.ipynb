## ЧАСТЬ 2 · Домашка урока 3 — три графика

Задание было: построить гистограмму, scatter и heatmap, и под каждым написать, что видно.

Подключаем библиотеки для графиков:

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

### График 1 — Гистограмма: «сколько чего?»
Смотрим распределение рейтинга зрителей.

```python
plt.figure(figsize=(8,4))
sns.histplot(data=df, x="AudienceScore", bins=20)
plt.title("Распределение рейтинга зрителей")
plt.xlabel("Рейтинг зрителей")
plt.ylabel("Количество фильмов")
plt.show()
```

**Что видно (пример хорошего вывода):**
> Большинство фильмов имеют рейтинг 50–70. Очень плохих (ниже 20) и идеальных (выше 90) мало.
> Основная масса — «средние» фильмы.

### График 2 — Scatter: «связаны ли два признака?»
Проверим: чем больше бюджет, тем больше сборы?

```python
plt.figure(figsize=(8,4))
sns.scatterplot(data=df, x="Budget", y="WorldGross")
plt.title("Бюджет и мировые сборы")
plt.xlabel("Бюджет фильма")
plt.ylabel("Мировые сборы")
plt.show()
```

**Что видно (пример вывода):**
> Точки в целом идут вверх: дорогие фильмы чаще собирают больше — есть положительная связь.
> Но есть исключения (выбросы): фильмы с большим бюджетом и слабыми сборами — провалы в прокате.

### График 3 — Heatmap: связь всех числовых признаков

```python
numeric = df.select_dtypes(include="number")

plt.figure(figsize=(8,6))
sns.heatmap(numeric.corr(), annot=True, cmap="coolwarm")
plt.title("Корреляция признаков")
plt.show()
```

**Что видно (пример вывода):**
> Самая сильная связь — между бюджетом и сборами (квадрат краснее остальных).
> Это логично: в дорогие фильмы вкладывают рекламу, их видит больше людей.
>
> **Связь или причина?** Большой бюджет помогает сборам, но не гарантирует их —
> провалы на scatter это подтверждают. Так что это связь, а не строгая причина.
