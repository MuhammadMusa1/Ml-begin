### Главная идея урока
- **Обычная программа:** человек пишет правила → программа выдаёт ответ.
- **Машинное обучение:** человек даёт примеры (данные) → машина сама находит правила.

## Шаг 1. Загружаем готовую модель
Её уже обучили на миллионах картинок — она знает 1000 категорий объектов.

```python
# pipeline — "готовый конвейер": модель + всё нужное в одной строке.
from transformers import pipeline

classifier = pipeline("image-classification",
                      model="google/vit-base-patch16-224")

print("Модель готова! Она знает 1000 категорий объектов.")
```

## Шаг 2. Загрузи свою картинку
Нажми кнопку **«Выбрать файл»** и выбери любое изображение со своего устройства.

```python
from google.colab import files
uploaded = files.upload()   # выбери картинку
```

## Шаг 3. Спроси у модели: что это?
Модель даст топ-3 догадки и насколько она в каждой уверена.

```python
from PIL import Image

image_name = list(uploaded.keys())[0]
img = Image.open(image_name)

predictions = classifier(img, top_k=3)

display(img)
print("\nМодель думает, что это:")
for p in predictions:
    print(f"  {p['label']} — уверена на {round(p['score']*100)}%")
```
