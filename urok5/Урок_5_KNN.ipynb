# Урок 5 — «Скажи мне, кто твой сосед» (KNN)

Сегодня ты обучишь **свою первую настоящую ML-модель** и она будет предсказывать!

**Идея KNN:** смотрим на K ближайших соседей нового объекта — какой класс у большинства, такой присваиваем.

Запускай ячейки по порядку: ▶ или **Shift+Enter**.

---
### Два обозначения на весь курс
- **X** — признаки, по которым предсказываем (возраст, пол, класс билета)
- **y** — ответ, который предсказываем (выжил / не выжил)

## Шаг 1. Загружаем данные
Берём знакомый Titanic и оставляем нужные столбцы.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# оставляем только нужные столбцы и убираем строки с пропусками
df = df[["Pclass", "Sex", "Age", "Fare", "Survived"]].dropna()
df.head()
```

## Шаг 2. Готовим данные для модели
Модель понимает только числа. Столбец `Sex` — это текст (male/female), превратим его в число.

```python
# male -> 1, female -> 0
df["Sex"] = (df["Sex"] == "male").astype(int)
df.head()
```

## Шаг 3. Делим на X и y
- **X** — признаки (по чему предсказываем)
- **y** — цель (что предсказываем)

```python
X = df[["Pclass", "Sex", "Age", "Fare"]]   # признаки
y = df["Survived"]                          # ответ

print("Признаки X:", list(X.columns))
print("Цель y: Survived")
```

## Шаг 4. Масштабируем признаки ⚖️
**Важно!** У KNN есть ловушка: `Fare` (0–500) намного крупнее `Age` (0–80).
Без выравнивания соседей будут выбирать почти только по цене билета — это нечестно.
`StandardScaler` приводит все признаки к одному масштабу.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)   # теперь все признаки в одном масштабе

print("Признаки масштабированы — теперь они равноправны.")
```

## Шаг 5. Обучаем модель KNN 🎉
Всего три строки — и у тебя обученная модель!

```python
from sklearn.neighbors import KNeighborsClassifier

# создаём модель: спрашиваем 5 ближайших соседей
model = KNeighborsClassifier(n_neighbors=5)

# обучаем на масштабированных данных
model.fit(X_scaled, y)

print("Модель обучена! Она готова предсказывать.")
```

## Шаг 6. Проверяем точность
`accuracy` — доля правильных ответов.

```python
from sklearn.metrics import accuracy_score

predictions = model.predict(X_scaled)
acc = accuracy_score(y, predictions)
print("Точность:", round(acc, 3))
```

⚠️ **Внимание:** мы проверили модель на тех же данных, на которых учили.
Это как списать с ответов! На следующем уроке узнаем, почему так нельзя и как правильно.

## Шаг 7. Предсказываем нового пассажира 🔮
Придумаем человека и спросим модель: выжил бы он?
Важно: нового пассажира тоже масштабируем тем же scaler.

```python
# [Pclass, Sex, Age, Fare]
# например: 1 класс, женщина (0), 25 лет, билет за 100
new_passenger = pd.DataFrame([[1, 0, 25, 100]],
                             columns=["Pclass", "Sex", "Age", "Fare"])

# масштабируем его так же, как обучающие данные
new_scaled = scaler.transform(new_passenger)

result = model.predict(new_scaled)
print("Предсказание:", "Выжил" if result[0] == 1 else "Не выжил")
```

## ⭐ Для тех, кто хочет больше — найди лучший K
Переберём разные значения K и сравним точность. Помни: бери нечётные K!

```python
for k in [1, 3, 5, 9, 15]:
    m = KNeighborsClassifier(n_neighbors=k)
    m.fit(X_scaled, y)
    acc = accuracy_score(y, m.predict(X_scaled))
    print(f"K={k}: точность {round(acc, 3)}")
```
