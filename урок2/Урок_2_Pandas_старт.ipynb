## Что такое Pandas и зачем он нужен

**Pandas** — это библиотека Python для работы с таблицами. Проще всего представить так:
**Pandas — это «Excel внутри Python»**.

Зачем он нужен:
- вручную пролистать тысячи строк невозможно, а Pandas отвечает на вопросы о данных за секунды;
- одной командой можно посчитать среднее, найти максимум, отфильтровать нужные строки;
- именно через таблицы Pandas мы будем кормить данные нашим будущим моделям.

Главный объект Pandas — **DataFrame** (таблица). Его принято называть `df`.

## Знакомимся с NumPy — математика для данных

Рядом с Pandas почти всегда работает **NumPy** (читается «нам-пай»).

Если Pandas — это «Excel в Python» (таблицы), то **NumPy — это супер-калькулятор для списков чисел**. Он умеет за одну операцию обработать сразу тысячи чисел.

Зачем он нужен:
- считает быстро и сразу со всем списком (а не по одному числу);
- на нём построены все ML-библиотеки и нейросети — внутри они «думают» массивами NumPy;
- сам Pandas внутри использует NumPy.

Главный объект NumPy — **массив** (array): список чисел, с которым можно работать как с единым целым.

## Шаг 7. NumPy в действии
`np` — короткое имя для numpy. Сравни: с обычным списком так нельзя, а с массивом — можно.

```python
import numpy as np

# оценки класса по контрольной
grades = np.array([5, 4, 3, 5, 4, 2, 5])

print("Средний балл:", grades.mean())
print("Самая высокая:", grades.max())
print("Самая низкая:", grades.min())
```

**Магия NumPy:** одна операция меняет сразу все числа.
Добавим каждому ученику 1 балл «за старание» — без цикла, одной строкой:

```python
# прибавляем 1 ко ВСЕМ оценкам сразу
print("Было: ", grades)
print("Стало:", grades + 1)
```
